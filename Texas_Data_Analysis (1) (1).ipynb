```python
def func(a):
    if a == 1:
        return 'CRCP'
    elif a == 2:
        return 'JCP'
    elif a == 3:
        return 'JCP'
    else:
        return "Flexible Pavement"

data["PAVEMENT TYPE"] = data['DETAILED PVMNT TYPE ROAD LIFE'].apply(func)
```

```python
def condition(a):
    if a == 'A - VERY GOOD 90-100':
        return 'Good or Better'
    elif a == 'B - GOOD 70-89':
        return 'Good or Better'
    elif a == 'C - FAIR 50-69':
        return 'Fair'
    elif a == 'D - POOR 35-49':
        return 'Poor'
    elif a == 'F - VERY POOR 1-34':
        return 'Very Poor'
    else:
        return None
    
data["PAVEMENT CONDITIONS"] = data['CONDITION SCORE CLASSIFICATION'].apply(condition)
```

```python
data['PAVEMENT CONDITIONS'].value_counts().plot.bar()
plt.title("Total data sections for each pavement condition")
plt.xlabel('Pavement condition')
plt.ylabel('Number of sections')
```

```python
plt.rcParams["figure.figsize"] = (10, 6)
order = ['Good or Better','Fair', 'Poor', 'Very Poor']
data.groupby('FISCAL YEAR')['PAVEMENT CONDITIONS'].value_counts().unstack().plot.bar(y=['Good or Better','Fair','Poor','Very Poor'])
plt.title("Total data section of different pavement conditions for each year")
plt.xlabel('Fiscal Year')
plt.ylabel('Number of section')
plt.xticks(rotation=0)
plt.legend(title ='Pavement Condition', bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (10, 6)
data.groupby('PAVEMENT CONDITIONS')['FISCAL YEAR'].value_counts().unstack().reindex(['Good or Better', 'Fair', 'Poor', 'Very Poor']).plot.bar()
plt.title("Total data section of different years for each pavement condition")
plt.xlabel('Pavement Condition')
plt.ylabel('Number of sections')
plt.xticks(rotation=0)
plt.show()
```
